```python
import pandas as pd
import numpy as np
```

```python
### Load in the data files
from google.colab import drive
drive.mount('/content/gdrive')
d1 = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/Bob and Mel.csv')
d2 = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/BobandMellCDM2.csv')
```

```python
d2.head()
```

```python
# split Time column to get minutes and seconds 
d2[['minute', 'sec']] = d2.Time.str.split(":", n=1, expand=True)

```

```python
# fixed the error at row 346 with :00 inserted into sec column.
#first find the problem
pd.set_option("display.max_rows", None, "display.max_columns", None)
#then cut off the offending portion
d2['sec'] = pd.Series(d2['sec']).str.replace(':00', '', regex=True)
```

```python
# convert time to milliseconds
time1 = d1[' timestamp']*1000
time2 = (pd.to_numeric(d2.minute)*60000+pd.to_numeric(d2.sec)*1000).astype(float)
```

```python
d1.insert(loc=0, column='millisec', value=time1)
d2.insert(loc=0, column='millisec', value=time2)
```

```python
DF = d1.merge(d2, how='outer', on='millisec').sort_values(by='millisec')
```

```python
# show columns of DF
DF.head()
```

```python
# to find the colomn where CDM starts
(DF.columns=='Session ID').nonzero()
```

```python
for r in range(1,len(DF)):
    
    # set previous row index
    
    p = r - 1
    
    # if there is no entry in df2 for a certain timepoint, use the previous entry
    # one entry stops when the next one begins
    
    if DF.iloc[r, 715:].isnull().all():
        
        DF.iloc[r, 715:] = DF.iloc[p, 715:]
    
    else:
        
        DF.iloc[r, 715:] = DF.iloc[r, 715:]
```
